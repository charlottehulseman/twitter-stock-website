```python
import pandas as pd
fields = {
    "logo_url": "logo",
    "website": "website",
    "industry": "industry",
    "longBusinessSummary": "summary", 
    "sector": "sector",
    "fullTimeEmployees": "size"
}

columns = ["name"]
columns.extend(fields.values())

def get_stock_metadata(stocks):
    stock_data = []
    for stock in stocks:
        ticker = yf.Ticker(stock)
        _data = [stock]
        for field in fields.keys():
            try:
                _data.append(ticker.info[field])
            except KeyError:
                _data.append("not listed")
        stock_data.append(_data)
    stock_data = pd.DataFrame(stock_data, columns=columns)
    return stock_data
```

```python
get_stock_metadata(["MSFT", "AAPL", "GOOG", "NVDA", "TSLA", "NFLX"]).to_csv()
```
